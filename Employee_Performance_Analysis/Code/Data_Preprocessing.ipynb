## Checking Skewness

```python
from scipy.stats import skew, kurtosis
def skewedness(data):
    skewed=[]
    for col in data:
        sk=data[col].skew()
        if(sk>=1.96 or sk<=-1.96):
            skewed.append(col)
    return skewed


print(f"\033[1mSkewed Column name : {skewedness(data_num)}")
print('\033[1mRange of skewness & kurtosis, S < |1.96|')
# Checking Kurtosis
print('\033[1mSkewness for YearsSinceLastPromotion:',skew(df['Years Since Last Promotion']))
print('\033[1mKurtosis for YearsSinceLastPromotion:',kurtosis(df['Years Since Last Promotion']))
```

## Label Encoding

```python
# Using Label Encoder converting Labeled data into Numerical data

from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()
final_df[['Gender', 'Education Background', 'Marital Status','Employees Department', 'Employees Job Role',
       'Business Travel Frequency', 'Over Time', 'Attrition']] = final_df[['Gender', 'Education Background', 
       'Marital Status','Employees Department', 'Employees Job Role','Business Travel Frequency', 'Over Time', 
       'Attrition']].apply(label.fit_transform)
```

```python
final_df
```

```python
# Defining Independent variable and Dependent variable as X & y
X = final_df.iloc[:,:-1]
y = final_df.iloc[:,-1]
X
```

### Checking the Dataset for Imbalanced Data

```python
y.value_counts(normalize=True)*100
```

* The data is imbalance. Now, We have to balance the data using RandomOverSampler technique.

```python
from imblearn.over_sampling import RandomOverSampler,SMOTE,SMOTENC

oversampling = RandomOverSampler()
X_smote, y_smote = oversampling.fit_resample(X,y)
X_smote = pd.DataFrame(X_smote)
X_smote.columns = ['Age', 'Distance From Home', 'Employee Hourly Rate',
       'Number of Companies Worked', 'Employee Last Salary Hike Percent',
       'Total Work Experience In Years', 'Training Times Last Year',
       'Experience Years At This Company', 'Experience Years In Current Role',
       'Years Since Last Promotion', 'Years With Curr Manager',
       'Employee Education Level', 'Employee Environment Satisfaction',
       'Employee Job Involvement', 'Employee Job Level',
       'Employee Job Satisfaction', 'Employee Relationship Satisfaction',
       'Employee Work Life Balance', 'Gender', 'Education Background',
       'Marital Status', 'Employees Department', 'Employees Job Role',
       'Business Travel Frequency', 'Over Time', 'Attrition']
X_smote
```
